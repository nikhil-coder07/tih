### Question 2: Subset the DataFrame (1 Marks)

Create a new DataFrame called `df_covid_subset` that contains only the following columns from the original `df_covid`: `Date_reported`, `Country`, `WHO_region`, `New_cases`, `New_deaths`. Display the first 5 rows of the new subset.

**Expected Output:** A table showing the first 5 rows with only the 5 specified columns.

```python
df_covid_subset = df_covid[
    ['Date_reported', 'Country', 'WHO_region', 'New_cases', 'New_deaths']
]

df_covid_subset.head()
```

### Question 3: Data Type Conversion (1 Mark)

The `Date_reported` column in `df_covid_subset` is currently an object (string). Convert this column to a proper datetime format. Then, print the `info()` of the DataFrame to confirm the change.

**Expected Output:** A summary from `.info()` showing that `Date_reported` is now of type `datetime64[ns]`.

```python
import datetime
df_covid_subset['Date_reported'] = pd.to_datetime(
    df_covid_subset['Date_reported']
)

df_covid_subset.info()
```

### Question 4: Filter by Date Range (2 Marks)

Create a new DataFrame called `df_covid_trimmed` by filtering `df_covid_subset` to include only the dates from '2020-03-01' to '2023-08-31', inclusive. Print the shape of the new trimmed DataFrame.

**Expected Output:** A tuple representing the shape of the filtered DataFrame (rows, columns).

```python
df_covid_trimmed = df_covid_subset[
    (df_covid_subset['Date_reported'] >= '2020-03-01') &
    (df_covid_subset['Date_reported'] <= '2023-08-31')
]

print(df_covid_trimmed.shape)
```

### Question 5: Find Top 5 Affected Countries (2 Marks)

From the original `df_covid` DataFrame, find the 5 countries with the highest total `Cumulative_cases`. Group the data by `Country`, find the maximum `Cumulative_cases` for each, sort the results, and get the top 5. Print the resulting Series.

**Expected Output:** A Pandas Series showing the top 5 countries as the index and their max cumulative cases as the values.

```python
top_5_countries = (
    df_covid.groupby('Country')['Cumulative_cases']
    .max()
    .sort_values(ascending=False)
    .head(5)
)

print(top_5_countries)
```

### Question 6: Calculate Daily Global Cases (2 Marks)

Using the `df_covid_trimmed` DataFrame, calculate the total number of `New_cases` for each date across all countries. Store this in a new DataFrame called `daily_global_cases` and print its head.

**Expected Output:** A Series showing the first few dates and the corresponding total new cases for that day.

```python
daily_global_cases = (
    df_covid_trimmed.groupby('Date_reported')['New_cases']
    .sum()
)

print(daily_global_cases.head())
```

### Question 7: Aggregate Data by Quarter (2 Marks)

Aggregate the `New_cases` and `New_deaths` from `df_covid_trimmed` on a quarterly basis. First, create a 'Quarter' column, then group by this column and sum the cases and deaths. Display the head of the resulting DataFrame.

**Expected Output:** A DataFrame with 'Quarter' as the index and columns for 'New_cases' and 'New_deaths'.

```python
df_covid_trimmed['Quarter'] = (
    df_covid_trimmed['Date_reported']
    .dt.to_period('Q')
)
quarterly_data = (
    df_covid_trimmed.groupby('Quarter')
    [['New_cases', 'New_deaths']]
    .sum()
)
quarterly_data.head()
```

### Question 8: Total Cases by WHO Region (2 Marks)

Calculate the total sum of `New_cases` for each `WHO_region` in the `df_covid_trimmed` DataFrame. Display the results sorted in descending order.

**Expected Output:** A Pandas Series showing each WHO region and their total new cases, sorted from highest to lowest.

```python
region_cases = (
    df_covid_trimmed.groupby('WHO_region')['New_cases']
    .sum()
    .sort_values(ascending=False)
)

print(region_cases)
```

### Question 9: Find the Day with the Highest Number of New Cases (1 Marks)

Using the `daily_global_cases` Series you created earlier, find the date on which the highest number of new cases was reported globally. Print the date and the corresponding number of cases.

**Expected Output:** A printout of the date with the highest cases and the number of cases on that day.

```python
max_date = daily_global_cases.idxmax()
max_cases = daily_global_cases.max()

print("Date:", max_date)
print("New Cases:", max_cases)
```

### Question 10: Pivot Table for Monthly Cases by Region (2 Marks)

Create a pivot table to show the total `New_cases` for each `WHO_region` aggregated by month. First, create a 'Month_Year' column. Then, create the pivot table with `WHO_region` as the index, `Month_Year` as columns, and the sum of `New_cases` as values. Display the resulting pivot table.

**Expected Output:** A table with regions as rows, month-years as columns, and aggregated new cases as the cell values.

```python
df_covid_trimmed['Month_Year'] = (
    df_covid_trimmed['Date_reported']
    .dt.to_period('M')
)

pivot_table = pd.pivot_table(
    df_covid_trimmed,
    values='New_cases',
    index='WHO_region',
    columns='Month_Year',
    aggfunc='sum',
    fill_value=0
)

pivot_table
```

### Question 11: Load MNIST Data (3 Marks)

Load the MNIST dataset using `sklearn.datasets.load_digits`. Separate the dataset into features (`X`) and target labels (`y`).
Print the shape of the features and the target arrays.

**Expected Output:** The shape of `X` and `y`.

```python
from sklearn.datasets import load_digits

digits = load_digits()

X = digits.data
y = digits.target

print("X shape:", X.shape)
print("y shape:", y.shape)
```

### Question 12: K-Means Clustering (5 Marks)

Import `KMeans` from `sklearn.cluster`. Initialize and fit the K-Means clustering model on the MNIST features (`X`).
Since we know there are 10 digits (0-9), set the number of clusters to 10. Assign the cluster labels to a variable `kmeans_labels`.

**Expected Output:** A successfully fitted K-Means model and the cluster labels array.

```python
from sklearn.cluster import KMeans

kmeans = KMeans(
    n_clusters=10,
    random_state=42,
    n_init=10
)

kmeans.fit(X)

kmeans_labels = kmeans.labels_

print(kmeans_labels)
```

### Question 13: F1 Score Evaluation for Clustering (5 Marks)

Evaluate the clustering performance using the F1 score. Since K-Means assigns arbitrary cluster labels, you will first need to map each cluster label to the most frequent true label in that cluster.
After mapping the labels, calculate and print the macro-averaged F1 score using `sklearn.metrics.f1_score`.

**Expected Output:** The calculated F1 score of the clustering.

```python
import numpy as np
from scipy.stats import mode
from sklearn.metrics import f1_score
mapped_labels = np.zeros_like(kmeans_labels)
for i in range(10):
    mask = (kmeans_labels == i)
    if np.sum(mask) > 0:
        mapped_labels[mask] = mode(
            y[mask],
            keepdims=True
        ).mode[0]
f1 = f1_score(
    y,
    mapped_labels,
    average='macro'
)
print("Macro F1 Score:", f1)
```
